```python
import math
#define the shape class
class Shape:
    def area(self):
        pass

    def perimeter(self):
        pass

    def colour(self):
        return "No colour specified"
#define circle class
class Circle(Shape):
    #set the fundamental
    def __init__(self, radius, colour="Red"):
        self.radius = radius
        self.colour_value = colour
    """
    find the area and perimeter using formula
    and set the colour
    """
    def area(self):
        return math.pi * self.radius**2

    def perimeter(self):
        return 2 * math.pi * self.radius

    def colour(self):
        return self.colour_value
#define rectangle class
class Rectangle(Shape):
    #set the fundamental
    def __init__(self, length, width, colour="Blue"):
        self.length = length
        self.width = width
        self.colour_value = colour
    """
    find the area and perimeter using formula
    and set the colour
    """
    def area(self):
        return self.length * self.width

    def perimeter(self):
        return 2 * (self.length + self.width)

    def colour(self):
        return self.colour_value
#define triangle class
class Triangle(Shape):
    #set the fundamental
    def __init__(self, side1, side2, side3, colour="Green"):
        self.side1 = side1
        self.side2 = side2
        self.side3 = side3
        self.colour_value = colour
    """
    finde the value of area using formula 
    if have the coulour input value return the value 
    else the fundamental corlour is green
    perimeter is add the value of 3 side
    """
    def area(self):
        s = (self.side1 + self.side2 + self.side3) / 2
        return math.sqrt(s * (s - self.side1) * (s - self.side2) * (s - self.side3))

    def perimeter(self):
        return self.side1 + self.side2 + self.side3

    def colour(self):
        return self.colour_value


def create_shape():
    #input the type of shape and colour
    shape_type = input("Enter shape type (Circle, Rectangle, Triangle): ").capitalize()
    color = input("Enter colour: ")
    """
    if shape type is circle input the radius
    and return the result
    if shape type is rectangle input the length and width
    and return the result
    if shape type is triangle input the 3 different side 
    and return the result 
    if the shape type is invalid type 
    print the error message
    """
    if shape_type == "Circle":
        radius = float(input("Enter radius: "))
        return Circle(radius, color)
    elif shape_type == "Rectangle":
        length = float(input("Enter length: "))
        width = float(input("Enter width: "))
        return Rectangle(length, width, color)
    elif shape_type == "Triangle":
        side1 = float(input("Enter side 1: "))
        side2 = float(input("Enter side 2: "))
        side3 = float(input("Enter side 3: "))
        return Triangle(side1, side2, side3, color)
    else:
        print("Invalid shape type. Please choose Circle, Rectangle, or Triangle.")
        return None
#print result
def calculator(shape):
    print(f"Area: {shape.area()}")
    print(f"Perimeter: {shape.perimeter()}")
    print(f"Colour: {shape.colour()}")
    print("\n")


#rest the implementation
shape = create_shape()

if shape:
    calculator(shape)
```

```python
class User:
    def __init__(self, username, email, password):
        # Constructor to initialize user attributes
        self.username = username
        self.email = email
        self.password = password

    def view_listings(self, listings):
        # Method to view available listings
        for listing in listings:
            print(f"{listing.title} - {listing.price}")

    def create_listing(self, item, listings):
        # Method for users to create and add an item for sale
        listings.append(item)
        print(f"Item '{item.title}' added successfully for sale.")

class Buyer(User):
    def __init__(self, username, email, password):
        # Constructor for Buyer, inheriting from User
        super().__init__(username, email, password)
        self.cart = []

    def add_to_cart(self, item):
        # Method for buyers to add items to their shopping cart
        self.cart.append(item)
        print(f"Item '{item.title}' added to the cart.")

    def purchase_items(self, listings):
        # Method for buyers to purchase items in their cart
        total_price = sum(item.price for item in self.cart)
        print(f"Purchased items for ${total_price}")
        self.cart = []

class Seller(User):
    def __init__(self, username, email, password):
        # Constructor for Seller, inheriting from User
        super().__init__(username, email, password)
        self.sales = []  # List to keep track of sales
        self.revenue = 0  # Accumulated revenue from sales

    def view_sales(self):
        # Method for sellers to view their sales and revenue
        for sale in self.sales:
            print(f"Sale: {sale.title} - {sale.price}")
        print(f"Total revenue: ${self.revenue}")

class Admin(User):
    def __init__(self, username, email, password):
        # Constructor for Admin, inheriting from User
        super().__init__(username, email, password)

    def approve_account(self, user):
        # Method for admin to approve a user account
        print(f"Approved account for {user.username}")

    def reject_account(self, user):
        # Method for admin to reject a user account
        print(f"Rejected account for {user.username}")

    def approve_listing(self, item):
        # Method for admin to approve a listing
        print(f"Approved listing for {item.title}")

    def reject_listing(self, item):
        # Method for admin to reject a listing
        print(f"Rejected listing for {item.title}")

class Item:
    def __init__(self, title, description, price):
        # Constructor to initialize item attributes
        self.title = title
        self.description = description
        self.price = price

# Example Usage:

# Create instances of users, items, and admin
buyer1 = Buyer("buyer1", "buyer1@email.com", "password")
seller1 = Seller("seller1", "seller1@email.com", "password")
admin = Admin("admin", "admin@email.com", "adminpassword")
item1 = Item("Laptop", "High-performance laptop", 1200.0)
item2 = Item("Smartphone", "Latest smartphone model", 800.0)

# Sellers add listings
seller1.create_listing(item1, seller1.sales)
seller1.create_listing(item2, seller1.sales)

# Buyers view listings
buyer1.view_listings(seller1.sales)

# Buyers add items to the cart
buyer1.add_to_cart(item1)
buyer1.add_to_cart(item2)

# Buyers purchase items
buyer1.purchase_items(seller1.sales)
seller1.revenue += sum(item.price for item in buyer1.cart)
seller1.sales.extend(buyer1.cart)

# Seller views sales
seller1.view_sales()

# Admin approves the seller's listing
admin.approve_listing(item1)
```
